```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
niter = [100,500,1000,2000,5000,10000]
cp1_t = [1.287527560000001,6.803036321999997,13.774026184000007,25.721109344999988,392.28631816699993,2471.1109853830003]
cp2_t = [1.2975619520000023,7.097315956000003,13.936859098,27.049076958,381.02909025100007,2471.1109853830003] # Change the final time
```

```python
plt.figure(figsize=(10,4))
plt.plot(niter,cp1_t,label = 'logsfr100myr', c = 'gold', lw = 2)
plt.plot(niter,cp2_t, label  = 'logsfr1gyr', ls = '--', c = 'k', lw = 2)
plt.plot(niter,np.add(cp1_t,cp2_t),label = 'Aggregate', c = 'r', lw = 2)
plt.xscale('log')
plt.title('get_SFR Computation Time vs. $N_{iteration}$')
plt.xlabel('$N_{iteration}$')
plt.ylabel('Elapsed Time [sec]')
plt.grid()
plt.legend(frameon = False)
plt.savefig('get_SFR_call_time.pdf', format = 'pdf')
plt.show()
```
